```python
import numpy as np
```

```python
# distance from origin - euclidean norm

# define an n-dimensional vector A

A = np.array([1,2,3,4,5])

# calculate the euclidean distance from the origin (L2 norm)

distance = np.linalg.norm(A)
```

```python
print("Euclidean distance from the origin : ", distance)
```

```python
# Euclidean distance in n dimensions
```

```python
# define 2 n-dimensional vectors A and B

A = np.array([1,2,3,4,5])
B = np.array([6,7,8,9,10])

# calculate the difference between 2 vectors

difference = A - B
```

```python
difference
```

```python
# calculate the Euclidean distance between A and B (L2 norm of the difference)

distance = np.linalg.norm(difference)

print(distance, "Euclidean distance between A and B")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean
```

```python
#1. Generate 5 random 3D vectors

vectors = np.random.rand(5,3)

vectors
```

```python
#2. Assign a random class (0 or 1) to each vector

classes = np.random.randint(0,2,5)

classes
```

```python
#3. plot the vectors on a 3D Matplotlib Graph

fig = plt.figure()

ax = fig.add_subplot(111, projection = "3d")

for vec,cls in zip(vectors, classes):
    ax.quiver(0,0,0, vec[0],vec[1], vec[2], color = ("r" if cls == 1 else "b"))
```

```python
# 4. get user input for a query point (3 D vector)
```

```python
query_vector = np.array(list(map(float,input("Enter the query point (3D vector seperated by space: ").split())))
```

```python
# 5. Calculate the distance from the query vector to the 5 vectors and find the nearest neighbours
```

```python
distances = [euclidean(query_vector, vec) for vec in vectors]
```

```python
nearest_neighbour_index = np.argmin(distances)
```

```python
nearest_neighbour_index
```

```python
# 6. output the class of the nearest neighbour

print("The class of the nearest neighbour is: ", classes[nearest_neighbour_index])
```

```python
# Generate random 2D data with 100 points
data = np.random.rand(100,2)
```

```python
data
```

```python
# Calculate the mean of the data along each dimension
```

```python
data_mean = np.mean(data, axis = 0)
```

```python
data_mean
```

```python
# perform mean centering
```

```python
centered_data = data - data_mean
```

```python
# uses of Dot product

# 1. compute similarity between 2 vectors
# 2. projection of vector B on vector A
# 3. perform matrix multiplication
```

```python
# Angle between 2 vectors
```

```python
# A.B = || A || || B || Cos0
```

```python
# Cosine Similarity - cos of angle between 2 vectors A and B

# Cos0 = A.B / ||A|| ||B||

# || A || - distance of A from Origin or magnitude of A
# || B || - distance of B from Origin or magnitue of B
```

```python
# define two vectors

A = np.array([1,2,3])
B = np.array([-4,-5,-6])
C = np.array([5,5,5])
```

```python
# calculate the cosine similarity

cosine_similarity = np.dot(A,B) / (np.linalg.norm(A) * np.linalg.norm(B))
```

```python
print("Cosine_similarity between A and B: ",cosine_similarity)
```

```python
cosine_similarity = np.dot(A,C) / (np.linalg.norm(A) * np.linalg.norm(C))
print("Cosine_similarity between A and C: ",cosine_similarity)
```
